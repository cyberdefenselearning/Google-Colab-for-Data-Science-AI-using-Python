Exercise 3:
Create a shopping cart and store the items in a dictionary

```python
items = ['eggs', 'bread', 'roll', 'cheese', 'milk', 'tomato', 'potato']  # Menu for the items
user_items = {}  # Dictionary to store user cart
```

```python
while True:
    print()

    # Show the menu and then the quit option
    for idx, i in enumerate(items):
        print (f'{idx+1}. {i}')

    print('Are you done, do you want to quit? press 0 at the next input.')

    # Get user selection
    user_selection = int(input('Choose the item, please enter he corresponding number (between 1 and 7): '))

    # Quit if the user enters 0, otherwise check and add the item to the dictionary
    if user_selection == 0:
        break

    if not(user_selection >= 1 and user_selection <= 7):
        print ('Invalid Selection, please make the correctio selection')
        continue

    # Ask for the number of items for selected item
    num_items = int(input(f'How many items for {items[user_selection-1]}: '))

    # Check of the items already exists in user cart, if yes then dont overwrite but add to it
    if items[user_selection-1] in user_items.keys():
        print ('Already there')
        user_items[items[user_selection-1]] += 1

    else:
        user_items[items[user_selection-1]] = num_items

```

```python
# Print the user cart
for k, v in user_items.items():
    print (f'{k} : {v} items')
```

Exercise 4:
Find the most frequent element in a list

```python
my_list = [4, 6, 2, 8, 10, 3, 2, 3, 2, 6, 8, 2, 9, 1, 0, 1, 1, 6, 9, 10, 3, 5, 7, 3, 3, 1]
```

```python
# Convert the list to a set to get unique elements, and convert back to list
unique_items = list(set(my_list))
```

```python
# Variables to track the max count
max_count = 0
max_number = 0

for n in unique_items:
    count = my_list.count(n)

    # Update the variables if the max_cout exceeds the previous one
    if count > max_count:
        max_count = count
        max_number = n

print (f'Most frequently occurring item in the list is {max_number}')
```
