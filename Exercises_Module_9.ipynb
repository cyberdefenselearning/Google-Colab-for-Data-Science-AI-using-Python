Exercise 2:<br>
Visualize the dataset and present the findings <br>

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Plot the NO2 against the Date to see concnetration change overtime - Line chart
fig1 = px.line(
    df,
    x='Date',
    y='NO2(GT)',
    title='NO2 concentration Over Time',
    labels = {'Date': 'Date', 'NO2(GT)' : 'NO2 Concentration (ug/m3)'},
    template='plotly_dark'
)

fig1.show()
```

Findings about the plot above:


*  Levels of NO2 fluctate over time
*  Levels of NO2 start to increase after Nov 2004



```python
# Distribution of CO levels - Histogram
fig2 = px.histogram(
    df,
    x='CO(GT)',
    nbins=20,
    title='Distribution of CO Levels',
    labels={'Date': 'Date', 'CO(GT)': 'CO Concentration (mg/m3)'},
    template='plotly_dark'
)

fig2.show()
```

Findings about the plot above:


*  Distribution of CO levels tell us that most prevalent is 1-1.9
*  Few outliers



```python
# Relationship between CO and NO2 - Scatterplot
fig3 = px.scatter(
    df,
    x='CO(GT)',
    y='NO2(GT)',
    title='Relationship between CO and NO2',
    labels={'CO(GT)': 'CO Concentration (mg/m3)', 'NO2(GT)' : 'NO2 Concentration (ug/m3)'},
    template='plotly_dark'
)

fig3.show()
```

Findings about the plot above:


*  Positive correlation between NO2 and CO
*  However, it is not a prefectly linear relationship
