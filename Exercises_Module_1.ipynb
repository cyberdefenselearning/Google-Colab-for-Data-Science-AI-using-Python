**Exercise 2**

Perform mathematical operations on 2 numbers given by the user


Steps to follow
1.   Take inputs as *float*
2.   Perform the operations:
        *   Addition
        *   Subtraction
        *   Multiplication
        *   Division
        *   Exponential
3. Format the output messages and print them

```python
first_num = float(input('Enter the first number: '))
second_num = float(input('Enter the second number: '))
```

```python
# Perform addition, subtraction, multiplication, division and exponential on the numbers from input
addition = first_num + second_num
subtraction = first_num - second_num
multiplication = first_num * second_num
division = first_num / second_num
exponential = first_num ** second_num
```

```python
# Formatting the message for the output, separately for each operation
message0 = f'Your input numbers are {first_num} and {second_num}'
message1 = f'Sum of the number is {addition}'
message2 = f'Difference of the number is {subtraction}'
message3 = f'Product of the numbers is {multiplication}'
message4 = f'Division of the numbers is {division}'
message5 = f'Exponential of the numbers is {exponential}'

print(message0, message1, message2, message3, message4, message5, sep='|', end='\n')
```
