Exercise 1:
Funtion to calculate the factorial of a number

```python
def factorial(num):
    result = 1  # Store the final result in this variable
    for i in range(num, 1, -1):
        result *= i

    return result
```

```python
output = factorial(5)
print (output)
```

Exercise 2:
Function to check if a number is prime or not

```python
def is_prime(num):
    prime = True
    num_divisibles = 0

    if num == 0 or num == 1 or num < 0:
        return False

    for i in range(1, int(num/2)+1):
        if num % i == 0:
            num_divisibles += 1

        if num_divisibles > 1:
            prime = False
            break

    return prime
```

```python
result = is_prime(17)
print (result)
```

Exercise 3:
Function to reverse the string

```python
def reverse_string(mystring):
    len_string = len(mystring)
    reversed_string = ''

    for i in range(len_string, 0, -1):
        reversed_string = reversed_string + mystring[i-1]

    return reversed_string
```

```python
test_string = 'Hello World'
output_string = reverse_string(test_string)
print (output_string)
```

Exercise 4:
Function to calculate area of shapes - circle, rectangle, and triangle

```python
def calculate_area(name, radius=None, length=None, height=None, base=None):
    """
        This function calculates the area of different shapes,
        It is better to use this function with keyword arguments instead
        of positional arguments to avoid any error
    """

    PI = 3.14
    area = 0

    # Check the name of the shape and then calculate area
    # Only calculate the area when the values are specified
    if name.lower() == 'circle':
        if radius is not None:
            area = PI * (radius ** 2)
        else:
            print ('Radius not specified')

    elif name.lower() == 'rectangle':
        if length is not None and height is not None:
            area = length * height
        else:
            print ('Length and Height not specified')

    elif name.lower() == 'triangle':
        if base is not None and height is not None:
            area = 0.5 * base * height
        else:
            print ('Base and Height not specified')

    else:
        print ('Incorrect shape name entered, please choose from circle, rectangle or triangle')

    return area
```

```python
myarea = calculate_area('triangle', base = 2, height = 4)
print (myarea)
```
