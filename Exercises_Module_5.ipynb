Exercise 1:
Create a class Dog with attributes name, breed, and age and methods bark() and fetch()

```python
class Dog:
    def __init__(self, n, b, a):
        self.name = n
        self.__breed = b
        self.age = a

    def get_name(self):
        return self.name

    def get_breed(self):
        return self.__breed

    def get_age(self):
        return self.age

    def set_name(self, n):
        self.name = n

    def set_age(self, a):
        if not (a < self.age):
            self.age = a
        else:
            print ('Invalid Age')

    def bark(self):
        print ('Barking Noise')

    def fetch(self):
        print ('Dog is fetching the ball')
```

```python
dog1 = Dog('Kelly', 'German Shephard', 2)
dog2 = Dog('Jonny', 'Poodle', 3)
```

```python
print (dog1.get_name(), dog1.get_breed(), dog1.get_age())
print (dog2.get_name(), dog2.get_breed(), dog2.get_age())
```

```python
dog1.set_age(3)
```

```python
dog1.get_age()
```

```python
dog2.set_age(1)
```

```python
dog1.bark()
dog2.fetch()
```

Exercise 2:
Create a class Cat which inherits from class Pet and demonstrate inheritance

```python
class Pet:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_name(self):
        return self.name

    def get_age(self):
        return self.age

    def set_name(self, n):
        self.name = n

    def set_age(self, a):
        if not (a < self.age):
            self.age = a
        else:
            print ('Invalid Age')

    def walk(self):
        print ('Cat is walking')

class Cat(Pet):
    def __init__(self, name, age, breed):
        super().__init__(name, age)
        self.__breed = breed

    def get_breed(self):
        return self.__breed

    def meow(self):
        print ('Cat is meowing')
```

```python
cat1 = Cat('Sally', 3, 'Persian')
cat2 = Cat('Kelly', 2, 'Persian')
```

```python
print (cat1.get_name(), cat1.get_breed(), cat1.get_age())
print (cat2.get_name(), cat2.get_breed(), cat2.get_age())
```

```python
cat1.walk()
cat2.walk()
```

```python
cat2.meow()
cat1.meow()
```

Exercise 3:
Implement a banking system with classes Bank, Customer and Account

```python
class Account:
    def __init__(self, balance=0):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amt):
        self.__balance += amt

    def withdraw(self, amt):
        if self.__balance >= amt:
            self.__balance -= amt
            print (f'Remaining Balance: {self.__balance}')

        else:
            print ('Insufficient funds')

class Customer:
    def __init__(self, id, name, add):
        self.__id = id
        self.name = name
        self.address = add
        self.account = Account()

    def get_name(self):
        return self.name

    def get_address(self):
        return self.address

    def set_address(self, add):
        self.address = add
        print (f'Address updated to {add}')

    def deposit(self, amt):
        self.account.deposit(amt)

    def withdraw(self, amt):
        self.account.withdraw(amt)

    def get_balance(self):
        return self.account.get_balance()

class Bank:
    def __init__(self, bn):
        self.bank_name = bn
        self.customers = {}

    def add_cusomter(self, id, n, add):
        self.customers[id] = Customer(id, n, add)

    def get_customer(self, id):
        return self.customers.get(id, 'None')

```

```python
bank = Bank('ABC Bank')
```

```python
bank.add_cusomter(1, 'Ben', 'XYZ Street')
```

```python
current_customer = bank.get_customer(1)
```
