Exercise 1:
Generator to create a fibonacci sequence

```python
def fibonacci():
    num1, num2 = 0, 1

    while True:
        yield num1
        num1, num2 = num2, num1 + num2
```

```python
fib = fibonacci()

print(next(fib))
print(next(fib))
print(next(fib))
print(next(fib))
print(next(fib))
```

```python
for _ in range(10):
    print (next(fib))
```

Exercise 2:
Decorator to log the function call with timestamp

```python
from datetime import datetime
# import functools

def log(func):
    def wrapper(*args, **kwargs):
        timestamp = datetime.now().strftime('%Y-%m-%d - %H:%M:%S')
        print (f'Calling the function: {func.__name__} - {timestamp}')

        result = func(*args, **kwargs)

        print (f'Finished the execution of function: {func.__name__}')
        return result
    return wrapper
```

```python
@log
def greet(name):
    print (f'Welcome {name}')

@log
def add(a, b):
    return a + b
```

```python
greet('John')
```

```python
res = add(4, 5)
```

Exercise 3:
Create a context manager to close the file, even if an error occurs

```python
import json

class FileManager:
    def __init__(self, fn, m):
        self.filename = fn
        self.mode = m

    def __enter__(self):
        self.file = open(self.filename, self.mode)
        return self.file

    def __exit__(self, exc_type, exc_value, traceback):
        if exc_type:
            print (f'Error encountered: {exc_value}')

        self.file.close()

        return False
```

```python
try:
    with FileManager('./sample_data/anscombe.json', 'r') as file:
        data = json.load(file)
        print (data)
except Exception as e:
    print (f'Handled the exception: {e}')
```
