Exercise 2:
Read a csv and find the average of a column

```python
import csv
```

```python
with open('./sample_data/california_housing_test.csv') as file:
    reader = csv.DictReader(file)

    sum = 0
    count = 0

    for row in reader:
        sum += float(row['housing_median_age'])
        count += 1

    print (f'Average of the column "housing_median_age" is {sum/count}')
```

Exercise 3:
Implement code to handle errors during numerical input

```python
def get_input():
    num = input('Please enter a number between 0 and 100: ')

    num = float(num)

    if type(num) != float:
        raise TypeError('Input is not a number')

    if float(num) < 0 or float(num) > 100 :
        raise ValueError('Number is less than 0 or greater than hundred')

    return num
```

```python
while True:
    try:
        num = get_input()
        break

    except (ValueError, TypeError) as e:
        print (e)
```

Exercise 4:
Implement a code to handle exception for invalid file formats - txt

```python
class InvalidFileFormatError(Exception):
    """
    Exception to be raised if input file format is not a txt
    """
    def __init__(self, fn):
        self.filename = fn
        super().__init__(f'Invalid file format, only txt is supported')
```

```python
try:
    path_file = './sample_data/california_housing_test.csv'

    if not path_file.endswith('.txt'):
        raise InvalidFileFormatError(path_file)

    print ('File is correct')
except InvalidFileFormatError as e:
    print (e)
```
