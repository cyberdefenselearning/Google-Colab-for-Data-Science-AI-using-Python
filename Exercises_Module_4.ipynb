Exercise 1:
Use numpy to perform matrix operation

```python
import numpy as np
```

```python
A = np.array([[3, 5], [7, 2]])
B = np.array([[1, 4], [6, 8]])
```

```python
# Addition
addition = A + B

print ('Addition of matrices A and B: ')
print (addition)
```

```python
# subtraction
subtraction = A - B

print ('Subtraction of matrices A and B: ')
print (subtraction)
```

```python
# Element wise multiplication
multiply = A * B

print ('Multiplication of matrices A and B: ')
print (multiply)
```

```python
# Dot product
dot_prodcut = np.dot(A, B)

print ('Dot product of matrices A and B: ')
print (dot_prodcut)
```

Exercise:
Create a line plot and a bar plot with matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_flights = sns.load_dataset('flights')
```

```python
df_flights
```

```python
# Group the data by year and compute maximum passengers
flights_tr = df_flights[['year', 'passengers']].groupby('year').max().reset_index()
flights_tr
```

```python
# Create a line plot
plt.figure(figsize=(8, 5))

plt.plot(flights_tr['year'], flights_tr['passengers'], marker = 'o', linestyle = '-', color='b', label='No. of Passengers')

plt.xlabel('Year')
plt.ylabel('Maximum Number of Passengers')
plt.title('Maximum number of passengers per year from 1949 to 1960')
plt.legend()

plt.show()
```

```python
# Bar graph
plt.figure(figsize=(10, 8))

plt.bar(flights_tr['year'], flights_tr['passengers'], color='skyblue', label='Passengers')
plt.xlabel('Year')
plt.ylabel('Maximum Number of Passengers')
plt.title('Maximum number of passengers per year from 1949 to 1960')
plt.legend()

plt.show()
```
